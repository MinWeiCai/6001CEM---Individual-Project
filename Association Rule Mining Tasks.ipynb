```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules, fpgrowth
from IPython.display import display

#Data import
file_path = "National_Health_and_Nutrition_Examination_Survey__NHANES__-_National_Cardiovascular_Disease_Surveillance_System.csv"
df = pd.read_csv(file_path)

#Data preprocessing
df_cleaned = df.drop(columns=["PriorityArea2", "PriorityArea4", "Geolocation"], errors='ignore')  #Remove Empty Columns
df_cvd = df_cleaned[df_cleaned["Class"] == "Cardiovascular Diseases"]  #Cardiovascular data only
df_selected = df_cvd[["Topic", "Question", "Data_Value"]].dropna()  #Selecting key fields & removing missing values

#Generate a more rational transactional dataset
df_selected['Feature'] = df_selected['Topic'] + ' - ' + df_selected['Question']
df_selected = df_selected.pivot_table(index='Feature', values='Data_Value', aggfunc='median')

#Set threshold, convert to boolean
threshold = df_selected['Data_Value'].median()
df_selected['Risk'] = df_selected['Data_Value'].apply(lambda x: x >= threshold)
df_selected = df_selected.drop(columns=['Data_Value'])

#Transaction format conversion
df_transactions = df_selected.T  #Make each variable a column
df_transactions = df_transactions.astype(bool)

#Checking transaction data for correctness
print("Sample transaction data:")
print(df_transactions.head())

#Apriori association rule mining
frequent_itemsets_apriori = apriori(df_transactions, min_support=0.01, use_colnames=True)
rules_apriori = association_rules(frequent_itemsets_apriori, metric="confidence", min_threshold=0.6)

#FP-Growth Association Rule Mining
frequent_itemsets_fpgrowth = fpgrowth(df_transactions, min_support=0.01, use_colnames=True)
rules_fpgrowth = association_rules(frequent_itemsets_fpgrowth, metric="confidence", min_threshold=0.6)

#Number of print rules
print(f"Number of rules discovered by Apriori: {len(rules_apriori)}")
print(f"Number of rules discovered by FP-Growth: {len(rules_fpgrowth)}")

#Visualisation - Apriori
plt.figure(figsize=(10, 6))
plt.barh(rules_apriori['antecedents'].astype(str), rules_apriori['confidence'])
plt.xlabel('Confidence')
plt.ylabel('Risk Factor Rules')
plt.title('Apriori: Confidence of Rules')
plt.show()

#Visualisation - FP-Growth
plt.figure(figsize=(10, 6))
plt.barh(rules_fpgrowth['antecedents'].astype(str), rules_fpgrowth['confidence'])
plt.xlabel('Confidence')
plt.ylabel('Risk Factor Rules')
plt.title('FP-Growth: Confidence of Rules')
plt.show()

#Show discovered rules
print("Top Apriori Rules by Confidence:")
display(rules_apriori.sort_values(by='confidence', ascending=False).head(10))

print("Top Apriori Rules by Lift:")
display(rules_apriori.sort_values(by='lift', ascending=False).head(10))

print("Top Apriori Rules by Support:")
display(rules_apriori.sort_values(by='support', ascending=False).head(10))

print("Top FP-Growth Rules by Confidence:")
display(rules_fpgrowth.sort_values(by='confidence', ascending=False).head(10))

print("Top FP-Growth Rules by Lift:")
display(rules_fpgrowth.sort_values(by='lift', ascending=False).head(10))

print("Top FP-Growth Rules by Support:")
display(rules_fpgrowth.sort_values(by='support', ascending=False).head(10))

#Top K Rules for Visual Lift Sorting
k = 10 

#Apriori-Top K by Lift
top_apriori_lift = rules_apriori.sort_values(by='lift', ascending=False).head(k)
plt.figure(figsize=(10, 6))
plt.barh(top_apriori_lift['antecedents'].astype(str), top_apriori_lift['lift'])
plt.xlabel('Lift')
plt.ylabel('Antecedents')
plt.title(f'Apriori: Top {k} Rules by Lift')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.figure(figsize=(12, 8))

#FP-Growth-Top K by Lift
top_fpgrowth_lift = rules_fpgrowth.sort_values(by='lift', ascending=False).head(k)
plt.figure(figsize=(10, 6))
plt.barh(top_fpgrowth_lift['antecedents'].astype(str), top_fpgrowth_lift['lift'])
plt.xlabel('Lift')
plt.ylabel('Antecedents')
plt.title(f'FP-Growth: Top {k} Rules by Lift')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.figure(figsize=(12, 8))
```
